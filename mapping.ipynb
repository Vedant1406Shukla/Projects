```python
import matplotlib.pyplot as plt
# line 1 points
x1 = [10,20,30]
y1 = [20,40,10]
# line 2 points
x2 = [10,20,30]
y2 = [40,10,30]

plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

# Display the figure.
plt.plot(x1,y1, color='b', linewidth = 3,  label = 'line1')
plt.plot(x2,y2, color='r', linewidth = 5,  label = 'line2')
plt.grid()
plt.xticks(color="red",size=15,rotation=10)
plt.yticks(color="green",size=12,rotation=-10)
plt.minorticks_on()
plt.tick_params(length=5,width=5,color="orange",axis='both',pad=20,direction='inout',which='major',labelright=False,labeltop=False,grid_alpha=1,grid_linestyle='--')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randn(1000)
y=np.sin(x)
z=np.cos(x)
plt.hist([x,y,z],bins="auto",ec="k",stacked=True)
plt.xlabel("age")
plt.ylabel("count of people")
plt.title("Population Count")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
Data=[[5,6,8], [10,11,14], [12,10,15]]

X=np.arange(3)
plt.plot(X,Data[0],color='b',label='Toy 1')
plt.plot(X,Data[1],color='g',label='Toy 2')
plt.plot(X,Data[2],color='r',label='Toy 3')
plt.legend(loc='upper left')
plt.title("Toy Sales",{"size":10,"color":"red"},backgroundcolor="yellow",loc="left",pad=0.9)
plt.xlabel("Toys",color="purple",size=20,backgroundcolor="yellow")
plt.ylabel("City Sales",color="orange",size=20,style="italic")
plt.minorticks_on()
plt.grid(color="black",which="minor")
plt.grid(True,color="red",which="major",axis="x")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,4*np.pi,0.5)
plt.plot(t,np.sin(t),marker="+",label="sin")
plt.plot(t,np.sin(t),marker="^",label="cos")
plt.title("sin and cos graph")
plt.xlabel("angle")
plt.ylabel("sin and cos value")
plt.legend(loc='upper right',ncol=1,bbox_to_anchor=(1,1))
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["USA","Russia","GBR","China","Germany"]

gold=[20,10,6,24,30]
silver=[100,50,30,120,150]
bronze=[200,100,60,240,300]

b_bronze=np.add(gold,silver)

plt.bar(x,gold,0.4,label="Gold")
plt.bar(x,silver,0.4,bottom=gold,label="silver")
plt.bar(x,bronze,0.4,bottom=b_bronze,label="bronze")

plt.xlabel("Countries")
plt.ylabel("medals")
plt.title("Medals Vs Countries")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
w=0.4
x=["CSE","ISE","ECE","EEE","Mech"]
boys=[100,40,60,240,300]
girls=[200,70,30,120,15]

plt.barh(x,boys,w,label="boys")
plt.barh(x,girls,w,left=boys,label="girls")
plt.xlabel("Courses")
plt.ylabel("Students")
plt.title("students Vs Courses")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
p= np.array([2,4,6])
x=np.arange(3)
plt.bar(x,p,color='r',width=0.3)
plt.bar(x+0.3,p*2,color='b',width=0.3)
plt.bar(x+0.6,p**2,color='g',width=0.3)
plt.title("Series Comparision")
plt.show()
```
